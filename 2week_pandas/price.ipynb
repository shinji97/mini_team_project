### 아파트 가격 분석

```python
# 모듈 로딩

import pandas as pd
```

```python
# 데이터 불러오기 & 확인
price = pd.read_excel('data/평균매매가격_종합.xlsx')

price.info()
```

```python
price.head(10)
```

```python
price_t = price.T

price_t.head(10)
```

```python
price_t.info()
```

```python
# 1~2번 행 결측치 확인
for i in range(1,3):
    print(f'{i}번째 행 : {price_t.iloc[i].isnull().sum()}')
```

```python
# 1~2번 행 삭제

price_t.drop(['Unnamed: 2','평균매매가격_종합'], inplace=True)
```

```python
# 0~8번 행 결측치 확인
for i in range(9):
    print(f'{i}번째 행 : {price_t[i].isnull().sum()}')
```

```python
# 0~8번 행 삭제
for j in range(9):
    price_t.drop(j,axis=1, inplace=True)

price_t
```

```python
price_t.rename({'* 통계표명 :':None}, inplace=True)
```

```python
# 9번 이라는 컬럼명을 '시점으로 변경'
price_t.rename({9:'시점'}, axis=1, inplace=True)

price_t.head()
```

```python
# 2017/10 ~ 22/9 필요한 데이터 추출

price_t[price_t['시점']=='2017년 10월']
```

```python
price_t[price_t['시점']=='2022년 09월']
```

```python
price_t_ne = price_t[70:130].copy(deep=True)

price_t_ne.head()
```

```python
price_t_ne.tail()
```

```python
# 시점 컬럼을 행 인덱스로 지정 & price_t 데이터 프레임의 0번 행을 컬럼 인덱스로 지정

price_t_ne.set_index('시점', inplace=True)
price_t_ne.columns = list(price_t.iloc[0])[1:]

price_t_ne.head()
```

```python
# 필요없는 행 삭제

price_t_ne.columns
```

```python
price_t_ne = price_t_ne[['전국','수도권','지방권','서울','경기','인천','부산','대구','광주','대전','울산','세종','강원','충북', '충남', '전북','전남','경북','경남','제주']]

price_t_ne.head()
```

```python
# 분기별로 묶기
# (2) 3개 행씩 평균

for i in range(0,60,3):
    price_t_ne.iloc[i:i+3] = round(price_t_ne.iloc[i:i+3].mean(),)

price_t_ne.head()
```

```python
# 중복된 행 제거 

price_t_ne.drop_duplicates(inplace=True)
```

```python
price_t_ne.head()
```

```python
price_t_ne.tail()
```

```python
# 필요한 값만 추출
price_result = price_t_ne.iloc[:,:3]
price_result

```

```python
# 인덱스 수정

idx = []
for i in range(17,23):
    for j in range(1,5):
        idx.append(f'{i}.{j}')

idx = idx[3:-1]
idx

price_result.index = idx
```

```python
price_result= price_result//100

price_result
```

### 아파트 가격 변동 원인 & 아파트 가격 비교

```python
# 아파트 거래량

ap_sell = pd.read_excel('data/아파트 합계 정리.xlsx')

ap_sell.set_index('시점', inplace=True)
ap_sell = ap_sell[['전국','수도권','지방']]

ap_sell.columns = ['아파트 거래량_전국', '아파트 거래량_수도권', '아파트 거래량_지방']


idx = []
for i in range(17,23):
    for j in range(1,5):
        idx.append(f'{i}.{j}')

idx = idx[3:-1]
idx

ap_sell.index = idx
ap_sell
```

```python
# 주택 매매량

ho_sell = pd.read_excel('data/주택 합계 정리.xlsx')

ho_sell
ho_sell.set_index('시점', inplace=True)
ho_sell = ho_sell[['전국','수도권','지방']]

ho_sell.columns = ['주택 매매량_전국', '주택 매매량_수도권', '주택 매매량_지방']


idx = []
for i in range(17,23):
    for j in range(1,5):
        idx.append(f'{i}.{j}')

idx = idx[3:-1]
idx

ho_sell.index = idx
ho_sell
```

```python
# 금리

rate = pd.read_excel('data/분기별 기준금리.xlsx')

rate['시점'] = rate['Unnamed: 0']

rate.set_index('시점', inplace=True)
rate = rate[['기준금리']]

rate
```

```python
# 가계 대출

home_money = pd.read_excel('data/통합 가계 대출.xlsx')


home_money['시점'] = home_money['Unnamed: 0']
home_money.set_index('시점', inplace=True)

home_money = home_money[['가계 대출_전국','가계 대출_수도권','가계 대출_지방']]

home_money
```

```python
# gdp

gdp = pd.read_excel('data/gdp.xlsx')

gdp.set_index('분기', inplace=True)

gdp = gdp[['경제성장률']]

gdp
```

```python
sell = pd.read_excel('data/착공_최종.xlsx')

sell['시점'] = sell['Unnamed: 0']
sell.set_index('시점', inplace=True)

sell = sell[['착공_전국','착공_수도권','착공_지방']]

sell
```

```python
not_sell = pd.read_excel('data/미분양_최종.xlsx')

not_sell['시점'] = not_sell['Unnamed: 0']
not_sell.set_index('시점', inplace=True)

not_sell = not_sell[['미분양_전국','미분양_수도권','미분양_지방']]

not_sell
```
