```python
# 모듈 로딩
import pandas as pd
```

```python
price_df = pd.read_csv('data/아파트_매매_실거래_평균가격_csv.csv', encoding='cp949')

price_df
```

```python
price_df.columns
```

```python
for i in range(1,6):
    price_df.drop(f'서울.{i}', axis=1, inplace=True)
    
price_df.head()
```

```python
price_df.head()
```

```python
price_df.columns
```

```python
price_df_column_list = list(price_df.iloc[0])

price_df_column_list[1] = '전국'
price_df_column_list[2] = '수도권'
price_df_column_list[3] = '지방'
price_df_column_list[11] = '세종'

price_df_column_list
```

```python
price_df.iloc[0] = price_df_column_list

price_df.iloc[0]
```

```python
addr_aliases = {'경기':'경기도', '경남':'경상남도', '경북': '경상북도', '충북': '충청북도',
                '서울':'서울특별시','부산':'부산광역시','대전':'대전광역시',
                '충남':'충청남도','전남':'전라남도','전북':'전라북도', '대구':'대구광역시',
                '인천':'인천광역시', '광주':'광주광역시', '울산':'울산광역시','세종':'세종특별자치시',
                '강원':'강원도', '제주':'제주특별자치도'}


price_df.columns = price_df.iloc[0].apply(lambda v: addr_aliases.get(v,v))

price_df.columns
```

```python
price_df
```

```python
price_df.drop(0, axis=0, inplace=True)

price_df.head()
```

```python
price_df['시점'] = price_df['시점'].str[:4]

price_df.rename(columns={'시점':'연도'}, inplace=True)

price_df.info()
```

```python
price_df.replace('-', 0, inplace=True)

price_df.info()
```

```python
price_df_column_list = list(price_df.columns)

price_df.iloc[:,1:] = price_df.iloc[:,1:].astype('float')

price_df.info()
```

```python
price_df = round(price_df.groupby('연도').mean(),1)

price_df
```

```python
not_need = ['전국', '수도권','지방','인천광역시','세종특별자치시','경기도','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도']

for col in not_need:
    price_df.drop(col, axis=1, inplace=True)
```

```python
price_df = price_df.T

price_df_index_list = list(price_df.index)
price_df.index = price_df_index_list


price_df_column_list = list(price_df.columns)
price_df.columns = price_df_column_list

price_df.head()
```

```python
price_df['평균'] = round(price_df.mean(axis=1),1)

price_df
```

```python
price_df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import platform

if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')

bar_width = 0.12

price_x = np.arange(0,len(price_df_column_list))

plt.figure(figsize=(20,5))
plt.bar(price_x - (bar_width*4),price_df.loc['서울특별시'][:-1], bar_width, label='서울특별시', color='#FFD8D8')
plt.bar(price_x - (bar_width*3),price_df.loc['부산광역시'][:-1] , bar_width, label='부산광역시', color = '#FAE0D4')
plt.bar(price_x - (bar_width*2),price_df.loc['대구광역시'][:-1] , bar_width, label='대구광역시', color='#FAECC5')
plt.bar(price_x - bar_width,price_df.loc['광주광역시'][:-1] , bar_width, label='광주광역시',color = '#F2CB61')
plt.bar(price_x,price_df.loc['대전광역시'][:-1] , bar_width, label='대전광역시', color = '#FAED7D')
plt.bar(price_x + bar_width,price_df.loc['울산광역시'][:-1] , bar_width, label='울산광역시', color = '#FFE08C')
plt.bar(price_x + (bar_width*2),price_df.loc['제주특별자치도'][:-1] , bar_width, label='제주특별자치도',color = '#FFC19E')
plt.bar(price_x + (bar_width*3),price_df.loc['강원도'][:-1] , bar_width, label='강원도', color= '#FFA7A7')
plt.xlabel('year')
plt.ylabel('price')
plt.legend()
plt.xticks(np.arange(0, 16 + (bar_width*4)), labels = price_df_column_list)
plt.title('아파트 매매 실거래 평균가격')

plt.savefig('data/아파트 매매 실거래 평균가격.png')
plt.show()
```

```python
color_list = ['#FFD8D8', '#FAE0D4','#FAECC5','#F2CB61','#FAED7D','#FFE08C','#FFC19E', '#FFA7A7']
plt.pie(price_df['평균'], labels=price_df_index_list, autopct='%.1f%%', colors=color_list)
plt.title('아파트 매매 실거래 평균가격 전국 비율')

plt.savefig('data/아파트 매매 실거래 평균가격 전국 비율.png')
plt.show()
```
